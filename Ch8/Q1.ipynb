#### Q1. What is the maximum memory-equivalent capacity of the following neural networks? <br>Assume binary classification, all weights are non-zero and all units have biases. There is enough information in the input that the first layer is not limited by it:



(a) The first deep neural network has the following capacity: <br>
1st layer = 4*3 = 12 <br>
2nd layer = min(12,3) = 3 <br>
3rd layer = min(4,3) = 3 <br>
Total = 12 + 3 + 3 = 18 bits <br>
<br>
(b) The second deep neural network has the following capacity: <br>
1st layer = 3 <br>
2nd layer = 4 (not limited to the output of the first neuron) <br>
3rd layer = 4 (not limited to the output of the second neuron) <br>
Total = 3 + 4 + 4 = 11 bits 

#### (c) What is the maximum amount of rows that each network in (a) and (b) can memorize?

The maximum amount of rows that (a) can memorize = MEC/log<sub>2</sub>(2) = 18/1 = 18 rows
The maximum amount of rows that (b) can memorize = MEC/log<sub>2</sub>(4) = 11/1 = 11 rows

#### (d) Answer (c) but for 4 classes instead of binary classification.

The maximum amount of rows that (a) can memorize for 4 classes = MEC/log<sub>2</sub>(4) = 18/1 = 9 rows
The maximum amount of rows that (b) can memorize for 4 classes = MEC/log<sub>2</sub>(4) = 11/1 = 5.5 = 5 rows
